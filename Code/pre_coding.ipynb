```python
import pandas as pd
import re
```

```python
business = pd.read_json('../Data/business_city.json', lines=True)
business.head()
```

```python
business.shape
```

## First, we want to see the number of businesses with each category.

```python
all_categories = []
for i in range(business.shape[0]):
    category = business.iloc[i,12]
    if category is not None:
        all_categories = all_categories + category.split(", ")
```

```python
len(all_categories)
```

The categories with number of businesses are shown below. Since the categories of "Restaurants" "Food" "Shopping" "NightLife" are too broad, and the categories of "Beauty & Spas" and "Home Services" are hard to understand, so I chose "Bars" as a pilot exploration.

```python
all_categories_count = {}
for element in all_categories:
   if element in all_categories_count:
      all_categories_count[element] += 1
   else:
      all_categories_count[element] = 1

{k: v for k, v in sorted(all_categories_count.items(), key=lambda item: -item[1])}
```

## Next, we're going to explore the sub_data.frame with "Bars" category
The next 2 chunks provides a subset of `business` dataframe. And then we obtain a data frame of 4374 rows.

```python
id_bars = []
for i in range(business.shape[0]):
    category = business.iloc[i,12]
    if category is not None:
        id_bars.append(re.search('Bar',category))
    else:
        id_bars.append(None)
```

```python
business_bars = business[[(x is not None) for x in id_bars]]
print(business_bars.shape)
business_bars.head()
```

```python
j = 0
for i in range(business_bars.shape[0]):
    if business_bars.iloc[i,11] is None:
        j+=1
print('There are ',j,'lines with no attributes!')
```

We'll find the attributes of the data.frame `business_bars`

```python
keys_bars = []
for i in range(business_bars.shape[0]):
    attribute = business.iloc[i,11]
    if attribute is not None:
        keys_bars = keys_bars +  list(attribute.keys())
```

```python
bars_attributes_count = {}
for element in keys_bars:
   if element in bars_attributes_count:
      bars_attributes_count[element] += 1
   else:
      bars_attributes_count[element] = 1

{k: v for k, v in sorted(bars_attributes_count.items(), key=lambda item: -item[1])}
```

Now, we define a function `extract_col` to extract certain attributes. And then we add the column with attributes to the data frame `business_bars`.

```python
def extract_col(key):
    col = []
    for attribute in business_bars.iloc[:,11]:
        if attribute is None:
            col.append(None)
        else:
            if key in attribute.keys():
                col.append(attribute[key])
            else:
                col.append(None)
    return(len(col),col)
```

```python
for key in {k: v for k, v in sorted(bars_attributes_count.items(), key=lambda item: -item[1])}.keys():
    business_bars[key] = extract_col(key)[1]
```
